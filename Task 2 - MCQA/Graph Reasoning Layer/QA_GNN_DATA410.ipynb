```python
import json
import random
import hashlib
from nltk.corpus import wordnet
from torch.utils.data import Dataset, random_split

def convert_format(question, choices, answer):
    unique_id = hashlib.md5(question.encode()).hexdigest()
    question_concept = question.split()[0]
    new_data = {
        "answerKey": answer,
        "id": unique_id,
        "question": {
            "question_concept": question_concept,
            "choices": [{"label": label, "text": choice} for label, choice in choices.items()],
            "stem": question
        }
    }
    return new_data

class MCQADataset(Dataset):
    def __init__(self, raw_data, augment_factor=0):
        self.data = []

        for item in raw_data:
            question = item['Question']
            choices = {label: item[label] for label in ['A', 'B', 'C', 'D']}
            choices['E'] = "None of the above"
            answer = item['Answer']
            original_data = convert_format(question, choices, answer)
            self.data.append(original_data)

            if augment_factor > 0:
                for _ in range(augment_factor):
                    augmented_question = self.augment_question(question)
                    augmented_data = convert_format(augmented_question, choices, answer)
                    self.data.append(augmented_data)

    def augment_question(self, question):
        words = question.split()
        idx_to_replace = random.randint(0, len(words) - 1)
        synonyms = [syn.lemmas()[0].name().replace('_', ' ') for syn in wordnet.synsets(words[idx_to_replace])]

        if synonyms:
            synonym = random.choice(synonyms)
            new_words = words.copy()
            new_words[idx_to_replace] = synonym
            return ' '.join(new_words)
        else:
            return question

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]

def load_data(file_path):
    with open(file_path, 'r') as file:
        return [json.loads(line) for line in file]

def save_to_jsonl(dataset, output_path):
    with open(output_path, 'w') as file:
        for idx in range(len(dataset)):
            item = dataset[idx]
            json.dump(item, file)
            file.write('\n')

raw_data = load_data('MMLU_train.jsonl')
train_size = int(0.8 * len(raw_data))
dev_size = len(raw_data) - train_size
train_raw_data, dev_raw_data = random_split(raw_data, [train_size, dev_size])

train_dataset = MCQADataset(train_raw_data, augment_factor=5)
dev_dataset = MCQADataset(dev_raw_data)
test_dataset = MCQADataset(load_data('MMLU_eval.jsonl'))

save_to_jsonl(train_dataset, 'train_rand_split.jsonl')
save_to_jsonl(dev_dataset, 'dev_rand_split.jsonl')
save_to_jsonl(test_dataset, 'test_rand_split_no.answers.jsonl')
```

```python
import json

def transform_data(input_file, output_file):
    transformed_data = []

    with open(input_file, 'r') as file:
        for line in file:
            item = json.loads(line)
            question = item["question"]["stem"]
            choices = item["question"]["choices"]
            answer_key = item["answerKey"]

            statements = []
            for choice in choices:
                statement_text = question.replace("?", " ") + choice["text"] + "."
                statements.append({
                    "label": choice["label"] == answer_key,
                    "statement": statement_text
                })

            transformed_item = {
                "answerKey": answer_key,
                "id": item["id"],
                "question": item["question"],
                "statements": statements
            }
            transformed_data.append(transformed_item)

    with open(output_file, 'w') as file:
        for item in transformed_data:
            json.dump(item, file)
            file.write('\n')

transform_data('train_rand_split.jsonl', './statement/train.statement.jsonl')
transform_data('dev_rand_split.jsonl', './statement/dev.statement.jsonl')
transform_data('test_rand_split_no.answers.jsonl', './statement/test.statement.jsonl')
```
